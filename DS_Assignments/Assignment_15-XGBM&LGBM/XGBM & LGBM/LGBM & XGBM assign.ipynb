##  Data Preprocessing

### 1.Impute missing values

```python
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
```

```python

test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

### 2. Encode categorical variables

```python
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python

test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)
```

## Building Predictive Models

### Split dataset into training and testing sets

```python
X_train = train_data.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'])
```

```python
y_train = train_data['Survived']
```

```python
X_test = test_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

## Build and train LightGBM model

```python
lgbm_model = LGBMClassifier()
lgbm_model.fit(X_train, y_train)

## Build and train XGBoost model

xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)

lgbm_pred = lgbm_model.predict(X_test)
xgb_pred = xgb_model.predict(X_test)

 #Comparative Analysis
# No ground truth available for test data, so we can't calculate metrics for test set
# We can only compare metrics on the training set
```

```python
# Choose evaluation metrics
def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return accuracy, precision, recall, f1

# Evaluate LightGBM model 
lgbm_metrics = evaluate_model(y_train, lgbm_model.predict(X_train))

# Evaluate XGBoost model
xgb_metrics = evaluate_model(y_train, xgb_model.predict(X_train))


# Compare performance metrics
print("LightGBM Metrics on Training Set:")
print("Accuracy:", lgbm_metrics[0])
print("Precision:", lgbm_metrics[1])
print("Recall:", lgbm_metrics[2])
print("F1-score:", lgbm_metrics[3])


print("\nXGBoost Metrics on Training Set:")
print("Accuracy:", xgb_metrics[0])
print("Precision:", xgb_metrics[1])
print("Recall:", xgb_metrics[2])
print("F1-score:", xgb_metrics[3])
```

```python
### # Performance comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
lgb_values = np.array(lgbm_metrics)
xgb_values = np.array(xgb_metrics)

# Bar plot for comparison
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, lgb_values, width, label='LightGBM')
rects2 = ax.bar(x + width/2, xgb_values, width, label='XGBoost')
```

```python
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

fig.tight_layout()
plt.show()

```
