### Removing outliers 

```python
z_scores = np.abs((data - data.mean()) / data.std())
data = data[(z_scores < 3).all(axis=1)]
```

###  Scaling the data

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# EDA
sns.pairplot(data)
plt.show()
```

## _Implementing Clustering Algorithms:_

### Hierarchy Clustering with evaluation metrics such as silhouette score 

```python
# create dendrogram
plt.figure(figsize = (20, 10));
dendrogram = sch.dendrogram(sch.linkage(data, method='average'))
```

```python
hierarchical = AgglomerativeClustering(n_clusters=5,linkage='ward')
hierarchical_labels = hierarchical.fit_predict(data_scaled)
print(f'Hierarchical Clustering Silhouette Score: {silhouette_score(data_scaled, hierarchical_labels)}')
```

```python
# save clusters for chart
y_hc = hierarchical.fit_predict(data)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines1 = pd.concat([data,Clusters], axis=1)
airlines1
```

```python
airlines1.sort_values("Clusters")
```

### K-means with evaluation metrics such as silhouette score 

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(data_scaled)
print(f'K-Means Silhouette Score: {silhouette_score(data_scaled, kmeans_labels)}')
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 19):
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 19), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### #DBSCAN with evaluation metrics such as silhouette score 

```python

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_scaled)
print(f'DBSCAN Silhouette Score: {silhouette_score(data_scaled, dbscan_labels)}')
```

### Visualization

```python
# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering')
```

```python
plt.subplot(1, 3, 2)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title('Hierarchical Clustering')
```

```python
plt.subplot(1, 3, 3)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')

plt.show()
```

### Cluster Analysis and Interpretation:

#### K-Means:

The clusters are spherical and evenly sized.

Suitable for datasets with distinct cluster boundaries.

#### Hierarchical:

Provides a hierarchy of clusters, which is useful for understanding the nested structure.

Different linkage methods can yield different cluster shapes.

#### DBSCAN:

Effectively identifies clusters of varying shapes and sizes.

Can detect noise points, making it robust to outliers.
