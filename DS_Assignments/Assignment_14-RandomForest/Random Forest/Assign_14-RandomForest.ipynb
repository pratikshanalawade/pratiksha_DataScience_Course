## 4: Random Forest Model Implementation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df.drop(['Type'], axis=1)

y = df['Type']
```

```python
# split data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 100)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

```python
from sklearn.metrics import accuracy_score

# instantiate the classifier 

rf = RandomForestClassifier(random_state=100)

# fit the model/Train the model

rf.fit(X_train, y_train)

# Predict the Test set results

y_pred = rf.predict(X_test)

# Check accuracy score 

print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# instantiate the classifier with n_estimators = 50

rfc_50 = RandomForestClassifier(n_estimators=50, random_state=100)



# fit the model to the training set

rfc_50.fit(X_train, y_train)



# Predict on the test set results

y_pred_50 = rfc_50.predict(X_test)



# Check accuracy score 

print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred_50)))
```

```python
# view the feature scores

feature_scores = pd.Series(rfc_50.feature_importances_, index=X_train.columns).sort_values(ascending=False)

feature_scores
```

```python
sns.barplot(x=feature_scores, y=feature_scores.index)

# Add labels to the graph
# Set the style
sns.set(style="darkgrid")
plt.xlabel('Feature Importance Score')

plt.ylabel('Features')

plt.title("Visualizing Features scores")

plt.show()
```

### 1. Explain Bagging and Boosting methods. How is it different from each other.

#### Bagging (Bootstrap Aggregating):

 Bagging is a method where multiple instances of a base learning algorithm are trained on different subsets of the training data. These subsets are typically created by sampling from the original dataset with replacement (bootstrap sampling). Each base model is trained independently of the others. When making predictions, Bagging aggregates the predictions of all base models, often by averaging (for regression) or voting (for classification).
rees.

Key characteristics of Bagging:

<b>Parallel Training:</b>
Base models are trained independently and in parallel.</n>

<b>Equal Weighting:</b></n>
All base models have equal weight in the final prediction.</n>

<b>Decreased Variance:</b></n>
Bagging reduces variance by averaging predictions from multiple models trained on different subsets of data, which helps in reducing overfitting.</n>

Examples: Random Forest is a popular ensemble learning algorithm based on Bagging, where the base models are decision trees.

#### Boosting:

Boosting is an iterative technique that adjusts the weights of observations based on the accuracy of previous models. In Boosting, each base model is trained sequentially, and at each iteration, the algorithm gives more weight to the training instances that were misclassified by the previous models. This focus on the "hard" examples allows subsequent models to improve their performance, gradually reducing the overall error. Boosting combines the predictions of all base models using a weighted sum.

### 2. Explain how to handle imbalance in the data.

Handling imbalanced data is crucial in machine learning, especially in classification tasks where one class significantly outnumbers the others. Failure to address class imbalance can lead to biased models that perform poorly on the minority class. Here are some common strategies to handle imbalanced data:

<b> 1 Resampling Techniques:</b>

Over-sampling: 
Increase the number of instances in the minority class by randomly duplicating samples or generating synthetic samples. Techniques like SMOTE (Synthetic Minority Over-sampling Technique) create synthetic samples based on the feature space.

Under-sampling: Decrease the number of instances in the majority class by randomly removing samples. This can be effective if the dataset is large and removing instances does not significantly impact the representativeness of the majority class.

<b> 2. Algorithmic Techniques:</b>

Class weights: Many algorithms allow you to assign higher weights to minority class instances during training. This makes the algorithm pay more attention to minority class samples.

Cost-sensitive learning: Adjust the misclassification cost for different classes. Misclassifying a minority class instance may incur a higher cost than misclassifying a majority class instance.

<b> 3. Evaluation Metrics:</b>

Use appropriate evaluation metrics: Instead of solely relying on accuracy, consider metrics like precision, recall, F1-score, or area under the ROC curve (AUC-ROC) that provide insights into the model's performance across different classes.

Confusion Matrix Analysis: Analyze the confusion matrix to understand how the model performs on each class individually, especially focusing on false positives and false negatives.

<b>4 Ensemble Methods:</b>

Ensemble methods: Techniques like Bagging and Boosting can be effective in handling class imbalance by combining multiple models. Boosting algorithms, in particular, tend to perform well with imbalanced data as they focus on difficult instances.

<b>5. Data Preprocessing:
</b

Feature Engineering: Carefully engineer features that help distinguish between different classes, especially for the minority class.

Normalization: Ensure that the features are properly scaled to prevent the model from being biased towards features with larger magnitudes.>
