```python
# Separate features and labels
features = data.drop('letter', axis=1)  # Assuming 'letter' is the label column
labels = data['letter']
```

```python
# Normalize the features
scaler = StandardScaler()
features_normalized = scaler.fit_transform(features)
```

```python
# Separate features and target variable
X = data.drop('letter', axis=1)
y = data['letter']
```

```python
# Display statistics of normalized features
print("Features Normalized Description:\n", pd.DataFrame(features_normalized, columns=features.columns).describe())
```

```python
# Encode the labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)
```

### 2. Model Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features_normalized, labels_encoded, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_train dtype:", X_train.dtype)
print("y_train dtype:", y_train.dtype)
```

```python
# TensorFlow session configuration (for GPU setup)
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    try:
        tf.config.experimental.set_memory_growth(physical_devices[0], True)
    except:
        pass  # Ignore errors if the memory growth is already set

# Construct a basic ANN model
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))  # Input layer + 1st hidden layer
model.add(Dense(64, activation='relu'))  # 2nd hidden layer
model.add(Dense(len(np.unique(y_train)), activation='softmax'))  # Output layer
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2)
```

```python
# Evaluate the model
y_pred = np.argmax(model.predict(X_test), axis=1)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

### 3. Hyperparameter Tuning

```python
from scikeras.wrappers import KerasClassifier
```

```python
#from sklearn.model_selection import GridSearchCV
```

```python
# Function to create model, required for KerasClassifier
def create_model(neurons=32, activation='relu', optimizer='adam',batch_size=16,epochs=50):
    model = Sequential()
    model.add(Dense(neurons, input_dim=X_train.shape[1], activation=activation))
    model.add(Dense(neurons, activation=activation))
    model.add(Dense(len(np.unique(y_train)), activation='softmax'))
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
 # Create the KerasClassifier wrapper
model = KerasClassifier(build_fn=create_model, verbose=0)
# Define the grid search parameters
param_grid = {
    'model__neurons': [32, 64, 128],
    'model__batch_size': [16, 32, 64],
    'model__epochs': [50, 100],
    'model__optimizer': ['SGD', 'Adam'],
    'model__activation': ['relu', 'tanh']
}

# Set up GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)

# Fit the grid search
grid_result = grid.fit(X_train, y_train)

# Summarize the results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))


# Evaluate the best model on the test set
best_model = grid_result.best_estimator_


# # Evaluate the best model on the test set
# y_pred = best_model.predict(X_test)
# if y_pred.ndim == 2:
#     y_pred = np.argmax(y_pred, axis=1)


# Predict using the best model found by grid search
y_pred_best = best_model.predict(X_test)

# Ensure the predictions are in the correct shape (1D array)
if y_pred_best.ndim == 2:
    y_pred_best = np.argmax(y_pred_best, axis=1)

# Print evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print("Classification Report:\n", classification_report(y_test, y_pred_best))
```
