### _Interview Questions:_

#### 1.	What is lift and why is it important in Association rules?

Lift is a measure used in association rule mining to evaluate the strength of a rule over the baseline probability of the co-occurrence of the items. In simpler terms, it helps determine how much more likely two items are to be bought together than if they were bought independently of each other.


#### Importance in Association Rules

#### Evaluating Strength of Rules:
        Lift helps in determining the strength of an association rule. A high lift value indicates a strong association between the items, which is valuable information for decision-making.


#### Filtering Rules:

        Lift can be used to filter out less interesting or less useful rules. For instance, rules with a lift value close to 1 might not be useful because they suggest no significant association between the items.


#### Understanding Dependencies:

        By analyzing lift values, one can understand the dependencies and relationships between different items in a dataset. This can be crucial for tasks such as market basket analysis, recommendation systems, and inventory management.

#### Beyond Simple Co-occurrence:

         While support and confidence provide insights into the frequency and reliability of a rule, lift goes a step further by considering the overall distribution of items. This makes it a more robust measure for assessing the true strength of an association.

#### What is support and Confidence. How do you calculate them?

#### 3.	What are some limitations or challenges of Association rules mining?

Association rule mining is a powerful technique for uncovering relationships between items in large datasets. However, it comes with several limitations and challenges that can affect its effectiveness and usability. Here are some of the key challenges:

#### 1. Scalability
Large Datasets: As the size of the dataset increases, the number of possible itemsets grows exponentially, leading to a combinatorial explosion. This makes the computation of support and confidence for all possible rules computationally intensive.
Memory and Time Constraints: Processing large datasets requires significant memory and time, which can be a limiting factor for practical applications.


#### 2. Threshold Setting
Support and Confidence Thresholds: Setting appropriate minimum support and confidence thresholds can be challenging. Too high thresholds might result in missing interesting rules, while too low thresholds can generate an overwhelming number of rules, many of which might be insignificant or spurious.


### 3. Interpretability
Too Many Rules: In practice, association rule mining often generates a large number of rules, many of which may be redundant or uninteresting. Identifying the most relevant and actionable rules among them can be difficult.
Complex Rules: The resulting rules can sometimes be complex and difficult to interpret, especially when involving multiple items and conditions.


#### 4. Rare Item Problem
Low Support Items: Items that occur infrequently (low support) may be part of interesting or important rules. However, they are often ignored if the support threshold is set too high.


#### 5. Actionability
Relevance to Business Objectives: Not all discovered associations are useful for business decisions. Determining the business relevance of the rules requires domain expertise and context-specific knowledge.
Implementation: Even if a rule is interesting, it might not always be actionable due to practical constraints or business policies.


#### 6. Data Quality and Preprocessing
Noise and Errors: Association rule mining is sensitive to noise and errors in the data. Inaccurate or incomplete data can lead to misleading rules.
Preprocessing Requirements: Effective rule mining often requires significant data preprocessing, such as data cleaning, handling missing values, and data transformation, which can be time-consuming.


#### 7. Overfitting
Spurious Rules: There is a risk of identifying patterns that are specific to the dataset but do not generalize to new data. This is often referred to as overfitting.


#### 8. Lack of Causality
Correlation vs. Causation: Association rules identify correlations but do not imply causation. Just because two items are frequently bought together does not mean one causes the other to be bought.


#### 9. Dynamic Data
Changing Patterns: In environments where data patterns change over time (e.g., seasonal trends), the discovered rules might become obsolete quickly, requiring frequent re-mining of the data.


#### 10. Evaluation Metrics
Beyond Support and Confidence: While support and confidence are commonly used metrics, they do not always capture the most interesting or useful rules. Measures like lift, leverage, and conviction can provide additional insights but also add complexity to the analysis.
